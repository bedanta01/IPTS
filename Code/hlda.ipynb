```python
dataset = pd.read_csv('../data/Labelled_tweets_v1.csv')
```

```python
tweets = dataset['Tweet']
```

```python
tweets = tweets.values.tolist()
```

```python
stopset = stopwords.words('english') + list(string.punctuation) + ['will', 'also', 'said','a','URL','USER_MENTION']
vocab =[]
```

```python
for i in range(0,len(tweets)):
    tweets[i] = tweets[i].split()
```

```python
for i in tweets:
    for j in i:
        if(j not in stopset):
            if(j not in vocab):
                vocab.append(j)
```

```python
vocab = sorted(list(vocab))
vocab_index = {}
for i, w in enumerate(vocab):
    vocab_index[w] = i
```

```python
tweets_after_stop_words=[]
for i in tweets:
    tweet = []
    for j in i:
        if(j not in stopset):
            tweet.append(j)
    tweets_after_atop_words.append(tweet)
```

```python
new_corpus = []
for doc in tweets_after_atop_words:
    new_doc = []
    for word in doc:
        #print(word)
        word_idx = vocab_index[word]
        new_doc.append(word_idx)
    new_corpus.append(new_doc)
```

```python
n_samples = 30       # no of iterations for the sampler
alpha = 10.0          # smoothing over level distributions
gamma = 1.0           # CRP smoothing parameter; number of imaginary customers at next, as yet unused table
eta = 0.1             # smoothing over topic-word distributions
num_levels = 3        # the number of levels in the tree
display_topics = 30   # the number of iterations between printing a brief summary of the topics so far
n_words = 10          # the number of most probable words to print for each topic after model estimation
with_weights = False  # whether to print the words with the weights
```

```python
hlda = HierarchicalLDA(new_corpus, vocab, alpha=alpha, gamma=gamma, eta=eta, num_levels=num_levels)
topicd = hlda.estimate(n_samples, display_topics=display_topics, n_words=n_words, with_weights=with_weights)
```
